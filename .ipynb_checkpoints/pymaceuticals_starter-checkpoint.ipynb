## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
merged_data
```

```python
# Checking the number of mice.
unique_mice = merged_data["Mouse ID"].unique()
len(unique_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate = merged_data[merged_data.duplicated(["Mouse ID", "Timepoint"])]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = merged_data.drop([merged_data.index[909], merged_data.index[911],
                               merged_data.index[913], merged_data.index[915],
                               merged_data.index[917]])
```

```python
# Checking the number of mice in the clean DataFrame.
unique_mice = clean_data["Mouse ID"].unique()
len(unique_mice)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
grouped_data = clean_data.groupby("Drug Regimen")
mean_tumor_size = grouped_data["Tumor Volume (mm3)"].mean()
median_tumor_size = grouped_data["Tumor Volume (mm3)"].median()
variance_tumor_size = grouped_data["Tumor Volume (mm3)"].var()
std_tumor_size = grouped_data["Tumor Volume (mm3)"].std()
sem_tumor_size = grouped_data["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.
tumor_data = pd.merge(mean_tumor_size, median_tumor_size, on="Drug Regimen", how="outer")
tumor_data = tumor_data.rename(columns={"Tumor Volume (mm3)_x": "Mean Tumor Volume (mm3)",
                                        "Tumor Volume (mm3)_y": "Median Tumor Volume (mm3)"})
tumor_data = pd.merge(tumor_data, variance_tumor_size, on="Drug Regimen", how="outer")
tumor_data = tumor_data.rename(columns={"Tumor Volume (mm3)": "Variance of Tumor Volume (mm3)"})
tumor_data = pd.merge(tumor_data, std_tumor_size, on="Drug Regimen", how="outer")
tumor_data = tumor_data.rename(columns={"Tumor Volume (mm3)": "Standard Deviation of Tumor Volume (mm3)"})
tumor_data = pd.merge(tumor_data, sem_tumor_size, on="Drug Regimen", how="outer")
tumor_data = tumor_data.rename(columns={"Tumor Volume (mm3)": "SEM of Tumor Volume (mm3)"})

tumor_data
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
grouped_data2 = clean_data.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
grouped_data2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_data = clean_data.groupby("Mouse ID").agg({"Timepoint": "max"})
greatest_data = greatest_data.rename(columns={"Timepoint": "Max Timepoint"})

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
greatest_data = pd.merge(greatest_data, clean_data, on="Mouse ID")
greatest_data = greatest_data.loc[greatest_data["Max Timepoint"] == greatest_data["Timepoint"]]
greatest_data = greatest_data.drop(columns="Max Timepoint")
greatest_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.     
# Locate the rows which contain mice on each drug and get the tumor volumes
treatment_data = greatest_data.loc[(greatest_data["Drug Regimen"] == treatments[0]) |
                                  (greatest_data["Drug Regimen"] == treatments[1]) |
                                  (greatest_data["Drug Regimen"] == treatments[2]) |
                                  (greatest_data["Drug Regimen"] == treatments[3])]
treatment_data = treatment_data[["Drug Regimen", "Tumor Volume (mm3)"]]

# create subsets
capomulin_data = treatment_data.loc[treatment_data["Drug Regimen"] == treatments[0]]
ramicane_data = treatment_data.loc[treatment_data["Drug Regimen"] == treatments[1]]
infubinol_data = treatment_data.loc[treatment_data["Drug Regimen"] == treatments[2]]
ceftamin_data = treatment_data.loc[treatment_data["Drug Regimen"] == treatments[3]]
    
# Determine outliers using upper and lower bounds
capomulin_quartiles = capomulin_data["Tumor Volume (mm3)"].quantile([.25,.5,.75])
ramicane_quartiles = ramicane_data["Tumor Volume (mm3)"].quantile([.25,.5,.75])
infubinol_quartiles = infubinol_data["Tumor Volume (mm3)"].quantile([.25,.5,.75])
ceftamin_quartiles = ceftamin_data["Tumor Volume (mm3)"].quantile([.25,.5,.75])

iqr = capomulin_quartiles[.75] - capomulin_quartiles[.25]
print(f"For Capomulin, the lower quartile of tumor volume is {capomulin_quartiles[.25]}, ")
print(f"        upper quartile of tumor volume is {capomulin_quartiles[.75]}, ")
print(f"        interquartile range of tumor volume is {iqr}, ")
print(f"        and median of tumor volume is {capomulin_quartiles[.5]}.")
print(f"        Values below {capomulin_quartiles[.25] - (1.5*iqr)} could be outliers.")
print(f"        Values above {capomulin_quartiles[.75] + (1.5*iqr)} could be outliers.")

capomulin_outliers = capomulin_data.loc[(capomulin_data["Tumor Volume (mm3)"] < 
                                        (capomulin_quartiles[.25] - (1.5*iqr))) |
                                        (capomulin_data["Tumor Volume (mm3)"] > 
                                        (capomulin_quartiles[.75] + (1.5*iqr)))]

iqr = ramicane_quartiles[.75] - ramicane_quartiles[.25]
print(f"For Ramicane, the lower quartile of tumor volume is {ramicane_quartiles[.25]}, ")
print(f"        upper quartile of tumor volume is {ramicane_quartiles[.75]}, ")
print(f"        interquartile range of tumor volume is {iqr}, ")
print(f"        and median of tumor volume is {ramicane_quartiles[.5]}.")
print(f"        Values below {ramicane_quartiles[.25] - (1.5*iqr)} could be outliers.")
print(f"        Values above {ramicane_quartiles[.75] + (1.5*iqr)} could be outliers.")

ramicane_outliers = ramicane_data.loc[(ramicane_data["Tumor Volume (mm3)"] < 
                                        (ramicane_quartiles[.25] - (1.5*iqr))) |
                                        (ramicane_data["Tumor Volume (mm3)"] > 
                                        (ramicane_quartiles[.75] + (1.5*iqr)))]                                 
                                         
iqr = infubinol_quartiles[.75] - infubinol_quartiles[.25]
print(f"For Infubinol, the lower quartile of tumor volume is {infubinol_quartiles[.25]}, ")
print(f"        upper quartile of tumor volume is {infubinol_quartiles[.75]}, ")
print(f"        interquartile range of tumor volume is {iqr}, ")
print(f"        and median of tumor volume is {infubinol_quartiles[.5]}.")
print(f"        Values below {infubinol_quartiles[.25] - (1.5*iqr)} could be outliers.")
print(f"        Values above {infubinol_quartiles[.75] + (1.5*iqr)} could be outliers.")

infubinol_outliers = infubinol_data.loc[(infubinol_data["Tumor Volume (mm3)"] < 
                                        (infubinol_quartiles[.25] - (1.5*iqr))) |
                                        (infubinol_data["Tumor Volume (mm3)"] > 
                                        (infubinol_quartiles[.75] + (1.5*iqr)))]                                        
                                       
iqr = ceftamin_quartiles[.75] - ceftamin_quartiles[.25]
print(f"For Ceftamin, the lower quartile of tumor volume is {ceftamin_quartiles[.25]}, ")
print(f"        upper quartile of tumor volume is {ceftamin_quartiles[.75]}, ")
print(f"        interquartile range of tumor volume is {iqr}, ")
print(f"        and median of tumor volume is {ceftamin_quartiles[.5]}.")
print(f"        Values below {ceftamin_quartiles[.25] - (1.5*iqr)} could be outliers.")
print(f"        Values above {ceftamin_quartiles[.75] + (1.5*iqr)} could be outliers.")

ceftamin_outliers = ceftamin_data.loc[(ceftamin_data["Tumor Volume (mm3)"] < 
                                        (ceftamin_quartiles[.25] - (1.5*iqr))) |
                                        (ceftamin_data["Tumor Volume (mm3)"] > 
                                        (ceftamin_quartiles[.75] + (1.5*iqr)))]

# Merge all outliers
outliers = pd.merge(capomulin_outliers, ramicane_outliers, on=["Drug Regimen", "Tumor Volume (mm3)"], how="outer")
outliers = pd.merge(outliers, infubinol_outliers, on=["Drug Regimen", "Tumor Volume (mm3)"], how="outer")
outliers = pd.merge(outliers, ceftamin_outliers, on=["Drug Regimen", "Tumor Volume (mm3)"], how="outer")

# Print potential outliers
print("")
print("Potential Outliers:")
print(outliers)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.figure()

# green_diamond to highlight our potential outliers
# From: https://matplotlib.org/stable/gallery/pyplots/boxplot_demo_pyplot.html
green_diamond = dict(markerfacecolor='g', marker='D')

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True)
fig.suptitle("Tumor Volume Across Regimens", fontsize=16, fontweight="bold")

tumor_vol = capomulin_data["Tumor Volume (mm3)"]
ax1.boxplot(tumor_vol, flierprops=green_diamond)
ax1.set_ylabel("Capomulin")

tumor_vol = ramicane_data["Tumor Volume (mm3)"]
ax2.boxplot(tumor_vol, flierprops=green_diamond)
ax2.set_ylabel("Ramicane")

tumor_vol = infubinol_data["Tumor Volume (mm3)"]
ax3.boxplot(tumor_vol, flierprops=green_diamond)
ax3.set_ylabel("Infubinol")

tumor_vol = ceftamin_data["Tumor Volume (mm3)"]
ax4.boxplot(tumor_vol, flierprops=green_diamond)
ax4.set_ylabel("Ceftamin")
plt.show()
```
